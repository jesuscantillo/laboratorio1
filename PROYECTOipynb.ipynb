# ***I-General Review Of Data***

*Student and applicant information. Learning path 1 is aimed at students in grades 9, 10 and 11 of official schools in the country. Students will be able to choose between programming fundamentals and digital creative industries. Learning path 2 is aimed at Colombians who are interested in developing programming skills. It has a basic cycle (basic programming, programming fundamentals and software development); and a deepening cycle (web application).*

```python
import pandas as pd
```

```python
file = "https://www.datos.gov.co/resource/2emd-i46m.csv"
df = pd.read_csv(file);
```

```python
df
```

**NaN values, these values will be cleaned.**

```python
#Elimination of columns with NaN values

df = df.drop(['meta_dinero'], axis=1)
df = df.drop(['meta_beneficiario'], axis=1)
df = df.drop(['inversion'], axis=1)
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#Limpieza de los estratos sociales "a" --- futura traduccion
df= df.drop(df[df["estrato_social"]=="a"].index)
```

```python
60 in df.edad.values
```

```python
121 in df.edad.values
```

```python
0 in df.edad.values
```

Possible bad data?
121 year old person?

```python

df_baddata_idx=df[df["edad"]==121].index
df = df.drop(df_baddata_idx)
df_baddata_idx=df[df["edad"]==0].index
df = df.drop(df_baddata_idx)
```
